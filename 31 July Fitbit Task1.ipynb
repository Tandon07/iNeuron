# 8 . Explore over an internet that how much calories burn is required for a healthy person and find out how many healthy person we have in our dataset 


```python
health_df=df.groupby(['Id'])[['Calories']].sum()
```

```python
health_df
```

```python
health_df['LoggedDays']=df.groupby('Id')['ActivityDate'].count().sort_values(ascending=False)
```

```python
health_df['healthIndex']=(health_df['LoggedDays']*2250)/health_df['Calories']

# 2250 is the daily aaverage calorie burn for healthy person 
```

```python
def healthStatus(a):
    if a<1:
        return "Healthy"
    else:
        return "Not Healthy"
```

```python
health_df
```

```python
health_df['healthyday']=health_df['healthIndex'].apply(healthStatus)
```
