# 9 . average weekly response time for each agent 

```python
def get_mins(time):
    h, m, s = time.split(':')
    return int(h)*60 + int(m) + int(s)/60
```

```python
df1
```

```python
t=[]
for i in df1['Average Response Time']:
    t.append(get_mins(i))
df1['Average weekly Response Time( in mins)']=t
```
